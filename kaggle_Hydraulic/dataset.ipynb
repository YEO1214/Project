```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from copy import deepcopy
import os
```

```python
file_list = os.listdir('./data')
file_list
```

```python
# get path to raw data
data_dir = os.path.abspath('./data')

# list of dataframes
file_df = {}
```

```python
for file in file_list:
    file_df[file[:-4]]=  pd.read_csv(f'{data_dir}/{file}', sep='\t', header=None)
```

```python
#target rename
file_df['profile'] = (
    file_df['profile'].rename({
        0:'cooler_eff', 1:'valve_perc', 2:'pump_leak',
        3:'accu_prs', 4:'stable'
    }, axis='columns'))
```

```python
file_df
```

```python
#평균 데이터
```

```python
tmp_mean = pd.DataFrame(columns = file_df.keys())
for i in file_df.keys():
    tmp_mean[i] = np.mean(file_df[i],axis=1)

tmp_mean = tmp_mean.drop('profile',axis=1)
tmp_mean
```

```python
data_mean = tmp_mean.join(file_df['profile'], how='right')
data_mean
```

```python
data_mean.to_csv('data_mean.csv',index=False)
```

```python
# 중앙값 데이터
```

```python
tmp_med = pd.DataFrame(columns = file_df.keys())
for i in file_df.keys():
    tmp_med[i] = np.median(file_df[i],axis=1)

tmp_med = tmp_med.drop('profile',axis=1)
tmp_med
```

```python
data_med = tmp_med.join(file_df['profile'], how='right')
data_med
```

```python
data_med.to_csv('data_median.csv',index=False)
```
